```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\oswal\Documents\Documentos\Ciencias de Datos\5Â° Semestre\Herramientas para grandes volumenes de datos\python\datos_abiertos_covid19\220120COVID19MEXICO.csv',
                 header=0,
                 parse_dates=['FECHA_ACTUALIZACION', 'FECHA_INGRESO', 'FECHA_SINTOMAS', 'FECHA_DEF'])
```

```python
df['SEXO'] = df['SEXO'].replace({1: 'MUJER',
                                 2: 'HOMBRE',
                                 3: 'NO ESPECIFICADO'})
```

## POSITIVOS

```python
#DATOS
#SEPARAMOS, UNIMOS Y AGRUPAMOS
df1 = df[(df['CLASIFICACION_FINAL'] == 1)]
df2 = df[(df['CLASIFICACION_FINAL'] == 2)]
df3 = df[(df['CLASIFICACION_FINAL'] == 3)]
full = pd.concat([df1, df2, df3])
plot = full.groupby(["FECHA_SINTOMAS", 'SEXO']).count()

#GRAFICO
sns.relplot(data=plot, x='FECHA_SINTOMAS', y='PAIS_NACIONALIDAD', hue='SEXO')
ax = plt.gca()
ax.axes.xaxis.set_visible(True)
ax.axes.yaxis.set_visible(True)
plt.xticks(rotation=25, ha='right')
```
